```
bus_logs
id	driver_id	time	diff
1	1	2018-06-14 12:00:44	0
2	2	2018-06-14 12:01:50	33
3	3	2018-06-14 12:04:20	-60
4	4	2018-06-14 12:06:28	5
5	1	2018-06-14 12:08:56	-10
6	2	2018-06-14 12:10:01	-1
7	3	2018-06-14 12:10:01	27
8	4	2018-06-14 12:10:01	30
9	1	2018-06-14 12:10:01	22
10	2	2018-06-14 12:10:01	64
11	3	2018-06-14 12:10:01	-18
12	4	2018-06-14 12:10:01	12
13	5	2018-06-14 12:10:01	30
14	2	2018-06-14 12:10:01	-40
15	3	2018-06-14 12:10:01	18
16	4	2018-06-14 12:10:01	10
17	1	2018-06-14 12:10:01	12
18	2	2018-06-14 12:10:01	34
19	3	2018-06-14 12:10:01	-42
20	4	2018-06-14 12:10:01	8
21	1	2018-06-14 12:10:01	5
22	2	2018-06-14 12:10:01	27
23	3	2018-06-14 12:10:01	21
24	4	2018-06-14 12:10:01	-4
25	5	2018-06-14 12:10:01	-30
```
Ответ: 
```
SELECT driver_id, ROUND(AVG(ABS(diff))) as avg_diff
FROM bus_logs
GROUP BY driver_id
HAVING avg_diff>=30
```

# 17. Лучшие сделки
В таблице deals содержится список сделок. Получите из таблицы следующие данные:

user_id — идентификатор пользователя;
deals — количество сделок сотрудника
sum_amount — сумма платежей по сделкам;
max_amount — максимальный платеж.
Учитывайте только закрытые сделки. Информацию выводите только для тех сотрудников, которые закрыли 3 и более сделок
```
deals

id	user_id	date	amount	status
1	1	2015-01-04	24500	closed
2	2	2015-01-04	62000	closed
3	3	2015-01-12	1350	closed
4	1	2015-01-14	600	new
5	2	2015-01-23	75000	closed
6	3	2015-01-31	35000	closed
7	4	2015-02-01	6800	closed
8	1	2015-02-16	12000	closed
9	2	2015-02-28	75000	new
10	3	2015-03-12	210000	closed
11	4	2015-03-12	35300	closed
12	1	2015-04-01	78000	closed
13	2	2015-04-14	112000	in_progress
14	3	2015-04-14	49000	closed
15	4	2015-04-14	84000	new
```
Ответ: 
```
SELECT user_id, COUNT(status) as deals, SUM(amount) as sum_amount, MAX(amount) as max_amount  FROM deals WHERE status='closed' GROUP BY user_id HAVING deals >=3
```
