`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–î–ê¬ª, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –∏ ¬´–ù–ï–¢¬ª ‚Äî –µ—Å–ª–∏ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª.

# 9. –í–æ–∑—Ä–∞—Å—Ç–Ω–∞—è –≥—Ä—É–ø–ø–∞

* –¥–æ 13 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ ‚Äì –¥–µ—Ç—Å—Ç–≤–æ;
* –æ—Ç 14 –¥–æ 24 ‚Äì –º–æ–ª–æ–¥–æ—Å—Ç—å;
* –æ—Ç 25 –¥–æ 59 ‚Äì –∑—Ä–µ–ª–æ—Å—Ç—å;
* –æ—Ç 60 ‚Äì —Å—Ç–∞—Ä–æ—Å—Ç—å.

# 23. –•–æ–¥ –∫–æ—Ä–æ–ª—è üå∂Ô∏è

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–µ—Ç—ã—Ä–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 8.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –®–∞—Ö–º–∞—Ç–Ω—ã–π –∫–æ—Ä–æ–ª—å —Ö–æ–¥–∏—Ç –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏, –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ –∏ –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –Ω–æ —Ç–æ–ª—å–∫–æ –Ω–∞ 1 –∫–ª–µ—Ç–∫—É.

```python
a, b, c, d = int(input()), int(input()), int(input()), int(input())
if -1 <= a - c <= 1 and -1 <= b - d <= 1:
    print('YES')
else:
    print('NO')
```

# 24. –ì–æ–Ω–∫–∞ —Å–ø–∏–¥—Å—Ç–µ—Ä–æ–≤
–ó—É–º –±—Ä–æ—Å–∏–ª –≤—ã–∑–æ–≤ –§–ª—ç—à—É –∏ –ø—Ä–µ–¥–ª–æ–∂–∏–ª –µ–º—É —á–µ—Å—Ç–Ω—ã–π –ø–æ–µ–¥–∏–Ω–æ–∫ –≤ –≤–∏–¥–µ –≥–æ–Ω–∫–∏ –≤–æ–∫—Ä—É–≥ –º–∞–≥–Ω–µ—Ç–∞—Ä–∞. –í —Å–ª—É—á–∞–µ –ø—Ä–æ–∏–≥—Ä—ã—à–∞ —ç—Ç–∞ –Ω–µ–π—Ç—Ä–æ–Ω–Ω–∞—è –∑–≤–µ–∑–¥–∞ –∑–∞—Ä—è–¥–∏—Ç—Å—è –∏ —É–Ω–∏—á—Ç–æ–∂–∏—Ç –º–∏—Ä, –ø–æ—ç—Ç–æ–º—É –§–ª—ç—à —Ä–µ—à–∏–ª –Ω–µ —Ä–∏—Å–∫–æ–≤–∞—Ç—å –±–µ–∑ –ø—Ä–∏—á–∏–Ω—ã, –∏ —É–∑–Ω–∞—Ç—å —É —Å–≤–æ–µ–≥–æ –¥—Ä—É–≥–∞ –¶–∏—Å–∫–æ –†–∞–º–æ–Ω–∞ –µ—Å—Ç—å –ª–∏ —Å–º—ã—Å–ª –ø—Ä–∏–Ω–∏–º–∞—Ç—å –≤—ã–∑–æ–≤. –¶–∏—Å–∫–æ –ø–æ–ª—É—á–∏–ª –¥–∞–Ω–Ω—ã–µ, —á—Ç–æ —Å–∫–æ—Ä–æ—Å—Ç—å –ó—É–º–∞ —Ä–∞–≤–Ω–∞ nn, –∞ —Å–∫–æ—Ä–æ—Å—Ç—å –§–ª—ç—à–∞ —Ä–∞–≤–Ω–∞ k.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –¶–∏—Å–∫–æ –Ω–∞ –≤–æ–ø—Ä–æ—Å –§–ª—ç—à–∞.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ nn –∏ kk, —Å–∫–æ—Ä–æ—Å—Ç—å –ó—É–º–∞ –∏ –§–ª—ç—à–∞.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ï—Å–ª–∏ –ó—É–º –±—ã—Å—Ç—Ä–µ–µ –§–ª—ç—à–∞ –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ ¬´NO¬ª, –µ—Å–ª–∏ –§–ª—ç—à –±—ã—Å—Ç—Ä–µ–µ –ó—É–º–∞ –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª, –µ—Å–ª–∏ –∏—Ö —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ä–∞–≤–Ω—ã –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ "Don't know".

```python
n = int(input())
k = int(input())
if n > k:
    print('NO')
elif n < k:
    print('YES')
else:
    print("Don't know")
```

# 25. –í–∏–¥ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤–∏–¥ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞, –¥–ª–∏–Ω—ã —Å—Ç–æ—Ä–æ–Ω –∫–æ—Ç–æ—Ä–æ–≥–æ —Ä–∞–≤–Ω—ã –≤–≤–µ–¥–µ–Ω–Ω—ã–º —á–∏—Å–ª–∞–º.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —á–∏—Å–ª–∞ ‚Äì –¥–ª–∏–Ω—ã —Å—Ç–æ—Ä–æ–Ω —Å—É—â–µ—Å—Ç–≤—É—é—â–µ–≥–æ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω —Ç–µ–∫—Å—Ç ‚Äì –≤–∏–¥ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ (¬´–†–∞–≤–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π¬ª, ¬´–†–∞–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–Ω—ã–π¬ª –∏–ª–∏ ¬´–†–∞–∑–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π¬ª).

```python
a, b, c = int(input()), int(input()), int(input())
if (a == b and a != c) or (b == c and b != a) or (a == c and a != b):
    print('–†–∞–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–Ω—ã–π')
elif a == b == c:
    print('–†–∞–≤–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π')
else:
    print('–†–∞–∑–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π')
```

# 26. –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ
–î–∞–Ω—ã —Ç—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ç—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –°—Ä–µ–¥–Ω–∏–º –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç –≤—Ç–æ—Ä—ã–º, –µ—Å–ª–∏ —Ç—Ä–∏ —á–∏—Å–ª–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è.

```python
a, b, c = int(input()), int(input()), int(input())
if c > a > b or b > a > c:
    print(a)
elif c > b > a or a > b > c:
    print(b)
else:
    print(c)
```

# 27. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π
–î–∞–Ω –ø–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä –º–µ—Å—è—Ü–∞ (1,2,‚Ä¶, 12). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ —ç—Ç–æ–º –º–µ—Å—è—Ü–µ. –ü—Ä–∏–Ω—è—Ç—å, —á—Ç–æ –≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –Ω–µ–≤–∏—Å–æ–∫–æ—Å–Ω—ã–º.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ü–æ—Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å –Ω–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, —Ç–∞–∫ —á—Ç–æ–±—ã –≤ –Ω–µ–π –±—ã–ª–æ –Ω–µ –±–æ–ª–µ–µ —Ç—Ä–µ—Ö —É—Å–ª–æ–≤–∏–π.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –ø–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä –º–µ—Å—è—Ü–∞.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ —ç—Ç–æ–º –º–µ—Å—è—Ü–µ.

```python
m = int(input())
if m == 1 or m == 3 or m == 5 or m == 7 or m == 8 or m == 10 or m == 12:
    print('31')
elif m == 2:
    print('28')
else:
    print('30')
```

# 28. –¶–µ—Ä–µ–º–æ–Ω–∏—è –≤–∑–≤–µ—à–∏–≤–∞–Ω–∏—è
–ò–∑–≤–µ—Å—Ç–µ–Ω –≤–µ—Å –±–æ–∫—Å–µ—Ä–∞-–ª—é–±–∏—Ç–µ–ª—è (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ). –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –≤–µ—Å —Ç–∞–∫–æ–≤, —á—Ç–æ –±–æ–∫—Å–µ—Ä –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç–Ω–µ—Å—ë–Ω –∫ –æ–¥–Ω–æ–π –∏–∑ —Ç—Ä–µ—Ö –≤–µ—Å–æ–≤—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π:

* –õ–µ–≥–∫–∏–π –≤–µ—Å ‚Äì –¥–æ 60 –∫–≥;
* –ü–µ—Ä–≤—ã–π –ø–æ–ª—É—Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å ‚Äì –¥–æ 64 –∫–≥;
* –ü–æ–ª—É—Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å ‚Äì –¥–æ 69 –∫–≥.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é, –≤ –∫–∞–∫–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –±—É–¥–µ—Ç –≤—ã—Å—Ç—É–ø–∞—Ç—å –¥–∞–Ω–Ω—ã–π –±–æ–∫—Å–µ—Ä.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç ‚Äì –Ω–∞–∑–≤–∞–Ω–∏–µ –≤–µ—Å–æ–≤–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏.

```python
w = int(input())
if w < 60:
    print('–õ–µ–≥–∫–∏–π –≤–µ—Å')
elif 60 <= w < 64:
    print('–ü–µ—Ä–≤—ã–π –ø–æ–ª—É—Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å')
elif 64 <= w < 69:
    print('–ü–æ–ª—É—Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å')
```

# 29. –°–∞–º–æ–ø–∏—Å–Ω—ã–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä üå∂Ô∏è

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ, –∏ —Å—Ç—Ä–æ–∫–∞.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –æ–ø–µ—Ä–∞—Ü–∏–∏ –∫ –≤–≤–µ–¥–µ–Ω–Ω—ã–º —á–∏—Å–ª–∞–º –∏–ª–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —Ç–µ–∫—Å—Ç, –µ—Å–ª–∏ –æ–ø–µ—Ä–∞—Ü–∏—è –Ω–µ–≤–µ—Ä–Ω–∞—è –ª–∏–±–æ –µ—Å–ª–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –ï—Å–ª–∏ —Å–º–µ—à–∞—Ç—å –∫—Ä–∞—Å–Ω—ã–π –∏ –∫—Ä–∞—Å–Ω—ã–π, —Ç–æ –ø–æ–ª—É—á–∏—Ç—Å—è –∫—Ä–∞—Å–Ω—ã–π –∏ —Ç.–¥.

# 31. –¶–≤–µ—Ç–∞ –∫–æ–ª–µ—Å–∞ —Ä—É–ª–µ—Ç–∫–∏ üå∂Ô∏è
–ù–∞ –∫–æ–ª–µ—Å–µ —Ä—É–ª–µ—Ç–∫–∏ –∫–∞—Ä–º–∞–Ω—ã –ø—Ä–æ–Ω—É–º–µ—Ä–æ–≤–∞–Ω—ã –æ—Ç 0 –¥–æ 36. –ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã —Ü–≤–µ—Ç–∞ –∫–∞—Ä–º–∞–Ω–æ–≤: 

* –∫–∞—Ä–º–∞–Ω 0 –∑–µ–ª–µ–Ω—ã–π;
* –¥–ª—è –∫–∞—Ä–º–∞–Ω–æ–≤ —Å 1 –ø–æ 10 –∫–∞—Ä–º–∞–Ω—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º –∏–º–µ—é—Ç –∫—Ä–∞—Å–Ω—ã–π —Ü–≤–µ—Ç, –∫–∞—Ä–º–∞–Ω—ã —Å —á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º ‚Äì —á–µ—Ä–Ω—ã–π;

* –¥–ª—è –∫–∞—Ä–º–∞–Ω–æ–≤ —Å 19 –ø–æ 28 –∫–∞—Ä–º–∞–Ω—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º –∏–º–µ—é—Ç –∫—Ä–∞—Å–Ω—ã–π —Ü–≤–µ—Ç, –∫–∞—Ä–º–∞–Ω—ã —Å —á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º ‚Äì —á–µ—Ä–Ω—ã–π;
* –¥–ª—è –∫–∞—Ä–º–∞–Ω–æ–≤ —Å 29 –ø–æ 36 –∫–∞—Ä–º–∞–Ω—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º –∏–º–µ—é—Ç —á–µ—Ä–Ω—ã–π —Ü–≤–µ—Ç, –∫–∞—Ä–º–∞–Ω—ã —Å —á–µ—Ç–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º ‚Äì –∫—Ä–∞—Å–Ω—ã–π.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –Ω–æ–º–µ—Ä –∫–∞—Ä–º–∞–Ω–∞ –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —ç—Ç–æ—Ç –∫–∞—Ä–º–∞–Ω –∑–µ–ª–µ–Ω—ã–º, –∫—Ä–∞—Å–Ω—ã–º –∏–ª–∏ —á–µ—Ä–Ω—ã–º. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ, –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –ª–µ–∂–∏—Ç –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –æ—Ç 0 –¥–æ 36.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ü–≤–µ—Ç –∫–∞—Ä–º–∞–Ω–∞ –ª–∏–±–æ —Å–æ–æ–±—â–µ–Ω–∏–µ ¬´–æ—à–∏–±–∫–∞ –≤–≤–æ–¥–∞¬ª, –µ—Å–ª–∏ –≤–≤–µ–¥—ë–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ª–µ–∂–∏—Ç –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –æ—Ç 0 –¥–æ 36.
![](https://st2.depositphotos.com/1072356/7725/v/950/depositphotos_77254344-stock-illustration-roulette-wheel.jpg)

```python
n = int(input())
if n == 0:
    print('–∑–µ–ª–µ–Ω—ã–π')
elif 1 <= n <= 10 or 19 <= n <= 28:
    if n % 2 == 0:
        print('—á–µ—Ä–Ω—ã–π')
    else:
        print('–∫—Ä–∞—Å–Ω—ã–π')
elif 11 <= n <= 18 or 29 <= n <= 36:
    if n % 2 != 0:
        print('—á–µ—Ä–Ω—ã–π')
    else:
        print('–∫—Ä–∞—Å–Ω—ã–π')
else:
    print('–æ—à–∏–±–∫–∞ –≤–≤–æ–¥–∞')
```

# 32. –ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ –æ—Ç—Ä–µ–∑–∫–æ–≤ üå∂Ô∏èüå∂Ô∏è
–ù–∞ —á–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ–π –¥–∞–Ω—ã –¥–≤–∞ –æ—Ç—Ä–µ–∑–∫–∞: [a1; b1] –∏ [a2; b2]. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –∏—Ö –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ.

–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ–º –¥–≤—É—Ö –æ—Ç—Ä–µ–∑–∫–æ–≤ –º–æ–∂–µ—Ç –±—ã—Ç—å:

* –æ—Ç—Ä–µ–∑–æ–∫;
* —Ç–æ—á–∫–∞;
* –ø—É—Å—Ç–æ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è 4 —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ a1,b1,a2,b2, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ a1 < b1 –∏ a2 <b2.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –≥—Ä–∞–Ω–∏—Ü—ã –æ—Ç—Ä–µ–∑–∫–∞, —è–≤–ª—è—é—â–µ–≥–æ—Å—è –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ–º, –ª–∏–±–æ –æ–±—â—É—é —Ç–æ—á–∫—É, –ª–∏–±–æ —Ç–µ–∫—Å—Ç ¬´–ø—É—Å—Ç–æ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ¬ª.

```python
a1, b1, a2, b2 = int(input()), int(input()), int(input()), int(input())
if a2 < a1:
    if b2 < a1:
        print('–ø—É—Å—Ç–æ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ')
    elif b2 == a1:
        print(b2)
    elif a1 < b2 <= b1:
        print(a1, b2)
    elif b2 > b1:
        print(a1, b1)
elif a2 == a1:
    if b2 <= b1:
        print(a2, b2)
    else:
        print(a2, b1)
elif a2 < b1:
    if b2 <= b1:
        print(a2, b2)
    else:
        print(a2, b1)
elif a2 == b1:
    print(a2)
else:
    print('–ø—É—Å—Ç–æ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ')
```
