–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –æ—à–∏–±–∫–∏ –≤ –∏–º–µ—é—â–µ–π—Å—è –ø—Ä–æ–≥—Ä–∞–º–º–µ, –∞ –Ω–µ –Ω–∞–ø–∏—Å–∞—Ç—å —Å–≤–æ—é, –≤–æ–∑–º–æ–∂–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É—é—â—É—é –¥—Ä—É–≥–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ—à–µ–Ω–∏—è.

````python
n = 8
count = 0
maximum = -10 ** 12
for i in range(1, n + 1):
    x = int(input())
    if x % 4 == 0:
        count += 1
        if x > maximum:
            maximum = x
if count > 0:
    print(count)
    print(maximum)
else:
    print('NO')
````

# 18. –í—Å–µ –≤–º–µ—Å—Ç–µ 2
–î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç:

* –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä 3 –≤ –Ω–µ–º;
* —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ –Ω–µ–º –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞;
* –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ—Ç–Ω—ã—Ö —Ü–∏—Ñ—Ä;
* —Å—É–º–º—É –µ–≥–æ —Ü–∏—Ñ—Ä, –±–æ–ª—å—à–∏—Ö –ø—è—Ç–∏;
* –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä, –±–æ–ª—å—à–∏—Ö —Å–µ–º–∏ (–µ—Å–ª–∏ —Ü–∏—Ñ—Ä –±–æ–ª—å—à–∏—Ö —Å–µ–º–∏ –Ω–µ—Ç, —Ç–æ –≤—ã–≤–µ—Å—Ç–∏ 1, –µ—Å–ª–∏ —Ç–∞–∫–∞—è —Ü–∏—Ñ—Ä–∞ –æ–¥–Ω–∞, —Ç–æ –≤—ã–≤–µ—Å—Ç–∏ –µ–µ);
* —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ –Ω–µ–º –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ü–∏—Ñ—Ä—ã 0 –∏ 5 (–≤—Å–µ–≥–æ —Å—É–º–º–∞—Ä–Ω–æ).

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö `

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

````python
n = int(input())
last = n % 10
counter_3 = 0
counter_ch = 0
total_5 = 0
total_7 = 1
counter_05 = 0
counter_last = 0
while n != 0:
    last_digit = n % 10
    if last_digit == 3:
        counter_3 += 1
    if last_digit % 2 == 0:
        counter_ch += 1
    if last_digit > 5:
        total_5 += last_digit
    if last_digit > 7:
        total_7 *= last_digit
    if last_digit == 5 or last_digit == 0:
        counter_05 += 1
    if last_digit == last:
        counter_last += 1
    n = n // 10
print(counter_3)
print(counter_last)
print(counter_ch)
print(total_5)
print(total_7)
print(counter_05)
````
