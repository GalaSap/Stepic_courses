`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö `

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –¥–ª—è –ø–µ—Ä–µ–≤–æ–¥–∞: C = 5/9 * (F - 32)

# 5. Dog age
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —á–∏—Å–ª–æ nn ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±–∞—á—å–∏—Ö –ª–µ—Ç. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–æ–∑—Ä–∞—Å—Ç —Å–æ–±–∞–∫–∏ –≤ —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏—Ö –≥–æ–¥–∞—Ö.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±–∞—á—å–∏—Ö –ª–µ—Ç.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–æ–∑—Ä–∞—Å—Ç —Å–æ–±–∞–∫–∏ –≤ —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏—Ö –≥–æ–¥–∞—Ö.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í —Ç–µ—á–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –ª–µ—Ç —Å–æ–±–∞—á–∏–π –≥–æ–¥ —Ä–∞–≤–µ–Ω 10.5 —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏–º –≥–æ–¥–∞–º. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –∫–∞–∂–¥—ã–π –≥–æ–¥ —Å–æ–±–∞–∫–∏ —Ä–∞–≤–µ–Ω 4 —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏–º –≥–æ–¥–∞–º.

```python
n = int(input())
if 1 <= n <= 2:
    n = n * 10.5
    print(n)
elif n > 2:
    n = 21 + (n - 2) * 4
    print(n)
```

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ü–µ–ª–æ–µ —Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
n = int(input())
n1 = n // 100
n2 = (n // 10) % 10
n3 = n % 10
a = (n1 + n2 + n3) - min(n1, n2, n3) - max(n1, n2, n3)
b = max(n1, n2, n3) - min(n1, n2, n3)
if a == b:
    print('–ß–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ')
else:
    print('–ß–∏—Å–ª–æ –Ω–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ')
```

# 12. –ê–±—Å–æ–ª—é—Ç–Ω–∞—è —Å—É–º–º–∞
–î–∞–Ω—ã –ø—è—Ç—å —á–∏—Å–µ–ª a1, a2, a3, a4, a5. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º—É –∏—Ö –º–æ–¥—É–ª–µ–π |a1| + |a2| +|a3| +|a4| + |a5|.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø—è—Ç—å –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª a1, a2, a3, a4, a5, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö` 

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì —Å—É–º–º—É –º–æ–¥—É–ª–µ–π –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö —á–∏—Å–µ–ª.

```python
a1 = float(input())
a2 = float(input())
a3 = float(input())
a4 = float(input())
a5 = float(input())
print(abs(a1) + abs(a2) + abs(a3) + abs(a4) + abs(a5))
```

# 20. –¢—Ä–∏ –≥–æ—Ä–æ–¥–∞
–î–∞–Ω—ã –Ω–∞–∑–≤–∞–Ω–∏—è —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å–∞–º–æ–µ –∫–æ—Ä–æ—Ç–∫–æ–µ –∏ —Å–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–∏—è —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å–∞–º–æ–µ –∫–æ—Ä–æ—Ç–∫–æ–µ –∏ –¥–ª–∏–Ω–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –¥–ª–∏–Ω—ã –Ω–∞–∑–≤–∞–Ω–∏–π –≤—Å–µ—Ö —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤ —Ä–∞–∑–ª–∏—á–Ω—ã.



```python
a, b, c = input(), input(), input()
if len(c) > len(a) < len(b):
    print(a)
elif len(c) > len(b) < len(a):
    print(b)
elif len(a) > len(c) < len(b):
    print(c)
if len(c) < len(a) > len(b):
    print(a)
elif len(c) < len(b) > len(a):
    print(b)
elif len(a) < len(c) > len(b):
    print(c)
```

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—Ç—Ä–æ–∫—É ¬´YES¬ª, –µ—Å–ª–∏ –∏–∑ –¥–ª–∏–Ω –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —Å–ª–æ–≤ –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—é, ¬´NO¬ª –≤ –∏–Ω–æ–º

# 26. –ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ
–ù–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏ (x1; y1) –∏ (x2; y2) –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–∞–∫ p = sqrt(x1 - x2)^2 + (y1 - y2)^2.
![](https://ucarecdn.com/26c7ba17-bee9-4f8f-9764-3b68b2ede790/)
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏, –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–¥–∞–Ω—ã.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —á–µ—Ç—ã—Ä–µ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ - x1, y1, x2, y2.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ.

```python
x1, y1, x2, y2 = float(input()), float(input()), float(input()), float(input())
from math import *
print(sqrt((pow((x1 - x2), 2)) + (pow((y1 - y2), 2))))
```

# 28. –°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è

* —Å—Ä–µ–¥–Ω–µ–µ –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b - sqrt(a * b)
* —Å—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b - 2ab / a + b
* —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ —á–∏—Å–µ–ª a –∏ b - sqrt(a^2 + b^2 / 2)

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ aa –∏ bb‚Äã, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ 4 —á–∏—Å–ª–∞ ‚Äì —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ, –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –∏ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ.

# 29. –¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â—É—é –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è sinx+cosx+tan^2x –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —á–∏—Å–ª—É –≥—Ä–∞–¥—É—Å–æ–≤ x.

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ xx –∏–∑–º–µ—Ä—è–µ–º–æ–µ –≤ –≥—Ä–∞–¥—É—Å–∞—Ö‚Äã. 

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏–Ω–∏–º–∞—é—Ç –∞—Ä–≥—É–º–µ–Ω—Ç –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö. –ß—Ç–æ–±—ã –ø–µ—Ä–µ–≤–µ—Å—Ç–∏ –≥—Ä–∞–¥—É—Å—ã –≤ —Ä–∞–¥–∏–∞–Ω—ã, –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ–æ—Ä–º—É–ª–æ–π r = x * pi / 180

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ú–æ–¥—É–ª—å math —Å–æ–¥–µ—Ä–∂–∏—Ç –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é radians(), –∫–æ—Ç–æ—Ä–∞—è –ø–µ—Ä–µ–≤–æ–¥–∏—Ç —É–≥–æ–ª –∏–∑ –≥—Ä–∞–¥—É—Å–æ–≤ –≤ —É–≥–æ–ª –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö.

```python
x = float(input())
from math import *
x1 = radians(x)
print(sin(x1) + cos(x1) + pow(tan(x1), 2))
```

# 31. –ö–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ üå∂Ô∏èüå∂Ô∏è
–î–∞–Ω—ã —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a, b, c. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ—Ä–Ω–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è ax^2 + bc + c = 0. 

`–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a != 0, b, c, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –µ—Å–ª–∏ –æ–Ω–∏ —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏–ª–∏ —Ç–µ–∫—Å—Ç ¬´–ù–µ—Ç –∫–æ—Ä–Ω–µ–π¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç –¥–≤–∞ –∫–æ—Ä–Ω—è, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ –∏—Ö –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è.

```python
a, b, c = float(input()), float(input()), float(input())
from math import *
d = pow(b, 2) - 4 * a * c
if a != 0 and d > 0:
    x1 = (-b - sqrt(d)) / (2 * a)
    x2 = (-b + sqrt(d)) / (2 * a)
    print(min(x1, x2))
    print(max(x1, x2))
elif a != 0 and d == 0:
    print(-b / (2 * a))
else:
    print('–ù–µ—Ç –∫–æ—Ä–Ω–µ–π')
```

# 32. –ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫
–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ ‚Äî –≤—ã–ø—É–∫–ª—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫, —É –∫–æ—Ç–æ—Ä–æ–≥–æ —Ä–∞–≤–Ω—ã –≤—Å–µ —Å—Ç–æ—Ä–æ–Ω—ã –∏ –≤—Å–µ —É–≥–ª—ã –º–µ–∂–¥—É —Å–º–µ–∂–Ω—ã–º–∏ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏. –ü–ª–æ—â–∞–¥—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Å –¥–ª–∏–Ω–æ–π —Å—Ç–æ—Ä–æ–Ω—ã aa –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å—Ç–æ—Ä–æ–Ω nn –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ: 
S = n * a^2 / 4tg(pi/n)

–î–∞–Ω—ã –¥–≤–∞ —á–∏—Å–ª–∞: –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ nn –∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ aa. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –ø–ª–æ—â–∞–¥—å —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

`–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö`
